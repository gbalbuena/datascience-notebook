```python
# misc
import os

# data
import pandas as pd

# database
import psycopg2
import sqlalchemy
from sqlalchemy import create_engine

# graphs
import matplotlib as plt
```

---

# Input

```python
POSTGRES_ADDRESS = os.getenv('POSTGRES_ADDRESS', 'localhost')
POSTGRES_PORT = os.getenv('POSTGRES_PORT', '5432') 
POSTGRES_USER = os.getenv('POSTGRES_USER', 'postgres')
POSTGRES_PASSWORD = os.getenv('POSTGRES_PASSWORD', 'postgres')
POSTGRES_DB = os.getenv('POSTGRES_DB', 'postgres')

print("database: " + POSTGRES_ADDRESS)
print("    port: " + POSTGRES_PORT)
print("  dbname: " + POSTGRES_DB)
print("username: " + len(POSTGRES_USER)*'*')
print("password: " + len(POSTGRES_PASSWORD)*'*')
```

# Database

## Establish connection

```python
# A long string that contains the necessary Postgres login information 
pg_connection_string = ('postgresql://{username}:{password}@{ipaddress}:{port}/{dbname}'.format(username=POSTGRES_USER, password=POSTGRES_PASSWORD, ipaddress=POSTGRES_ADDRESS, port=POSTGRES_PORT, dbname=POSTGRES_DB))

cnx = create_engine(pg_connection_string)
pg_connection_string
```

## Retrieve data

```python
pd.read_sql_query('''select * from information_schema.tables;''', cnx)
```
